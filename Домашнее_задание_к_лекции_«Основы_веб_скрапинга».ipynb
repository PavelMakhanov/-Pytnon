```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import time

def search_posts_habr(query):
    search_habr = pd.DataFrame()
    for q in query:
        URL = 'https://habr.com/ru/search/'
        params = {
            'q': q
        }
        req = requests.get(URL, params=params)
        time.sleep(0.3)
        soup = BeautifulSoup(req.text)
        search = soup.find_all('article', class_='tm-articles-list__item')
        try:
            for el in search:
                title = el.find('a', 'tm-article-snippet__title-link').text
                link ='https://habr.com'+el.find('a', 'tm-article-snippet__title-link').get('href')
                date = el.find('time').get('title')
                row = {'date': date, 'title': title, 'link': link}  
                search_habr = pd.concat([search_habr, pd.DataFrame([row])])  
        except AttributeError:
          pass       
    return search_habr.reset_index(drop=True)
```

```python
search_posts_habr(['python', 'анализ данных'])
```

```python
import time

def all_search_posts_habr(query,pages):
    all_search_posts_habr = pd.DataFrame()
    for i in range(1,pages+1):
        for q in query:
            URL = 'https://habr.com/ru/search/page'+str(i)
            params = {
                'q': q
            }
            req = requests.get(URL, params=params)
            time.sleep(0.3)
            soup = BeautifulSoup(req.text)
            search = soup.find_all('article', class_='tm-articles-list__item')
            try:
                for el in search:
                    title = el.find('a', 'tm-article-snippet__title-link').text
                    link ='https://habr.com'+el.find('a', 'tm-article-snippet__title-link').get('href')
                    url_text = requests.get(link)
                    soup_ = BeautifulSoup(url_text.text)
                    text = soup_.find('div', 'tm-article-body').text
                    date = el.find('time').get('title')
                    like = el.find('span', 'tm-votes-meter__value tm-votes-meter__value tm-votes-meter__value_positive tm-votes-meter__value_appearance-article tm-votes-meter__value_rating').get('title')
                    row = {'date': date, 'title': title, 'link': link, 'text': text, 'like':like}  
                    all_search_posts_habr = pd.concat([all_search_posts_habr, pd.DataFrame([row])])  
            except AttributeError:
              pass       
    return all_search_posts_habr.reset_index(drop=True)
```

```python
all_search_posts_habr(['python', 'анализ данных'],3)
```
