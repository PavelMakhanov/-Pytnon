##Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

**Считываем файлы и создаем датафреймы. Объединяем их и создаем датафрейм с названием фильма и оценкой**

```python
ratings = pd.read_csv('u.data.csv', sep = '\t', names = ["user id", "item id" , "rating" , "timestamp"])
```

```python
ratings.head()
```

```python
movies=pd.read_csv('u.item.csv', sep="|", encoding='ISO-8859-1', names= ["movie id", "movie title", "release date" ,
              "video release date", "IMDb URL" , "unknown" , "Action" , "Adventure" , "Animation",
              "Children's", "Comedy" , "Crime" , "Documentary" , "Drama" , "Fantasy",
              "Film-Noir", "Horror" , "Musical" , "Mystery" , "Romance" , "Sci-Fi",
              "Thriller" , "War" , "Western"])
```

```python
movies.head()
```

```python
df_merge= ratings.merge(movies, left_on='item id', right_on='movie id', how='inner')
```

```python
df_merge.head()
```

```python
movies_ratings= df_merge[['movie title','rating']]
```

```python
movies_ratings.head()
```

**Создаем датафрейм в которых оценка 5**

```python
movies_ratings_5 = movies_ratings[movies_ratings['rating']==5]
```

```python
movies_ratings_5.head()
```

**Считаем количество оценок и сортируем по убыванию**

```python
df_count=movies_ratings_5.groupby('movie title').count().sort_values(by='rating',ascending=False)
```

```python
df_count.head()
```

##Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
```

```python
power.head()
```

**Фильтруем по странам согласно условию**

```python
power_contry=power[(power['country']=='Latvia')|(power['country']=='Lithuania')|(power['country']=='Estonia')]
```

```python
power_contry.head()
```

**Фильтруем по категориям согласно условию**

```python
power_category = power_contry[(power_contry['category'] == 4)|(power_contry['category'] == 12)|(power_contry['category'] == 21)]
```

```python
power_category.head()
```

**Фильтруем по годам согласно условию**

```python
power_year = power_category[power_category['year'].between(2005, 2010)]
```

```python
power_year.head()
```

**Значение quantity должно быть положительным, поэтому отфильтруем это значение**

```python
power_quantity =power_year[power_year['quantity'] > 0]
```

```python
power_quantity.head()
```

**Считаем сумму потребления**

```python
sum(power_quantity['quantity'])
```
